```python
#异常值检测
import numpy as np

def detect_outliers_zscore(data, threshold=3):
#   使用Z分数法检测异常值
#    data: 输入的数据数组
#    threshold: 阈值，超过该阈值被视为异常值，默认为3
#   返回一个布尔型数组，表示每个数据点是否为异常值
 
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    outliers = np.abs(z_scores) > threshold
    return outliers

data = np.array([1, 2, 3, 4, 5, 20])
outliers = detect_outliers_zscore(data)
print(outliers)
```

```python
#特征选择
corr = data.corr()
print(corr['AQI'].sort_values(ascending=False))

X = data[['SO2','NO2','CO','PM2.5','PM10']]
y = data['AQI']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.3,random_state=42)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#模型选择与评估(此次采用线性回归模型，因为在天气预测中对子温度这一类连续性变量，能够准确地
#预测未来天气数据的趋势，而且线性回归模型也是比较简单易僮的) 

# 线性回归模型
lr = LinearRegression()
scores = cross_val_score(lr, X_train, y_train, cv=5)

# 交叉验证评估
print("交叉验证R^2得分:", np.mean(scores))

# 模型拟合
lr.fit(X_train, y_train) # 使用训练数据集对模型进行拟合

# 测试集评估
y_pred = lr.predict(X_test)  # 使用测试数据集进行预测
print("测试集MSE得分:", mean_squared_error(y_test, y_pred))  # 均方误差（MSE）
print("测试集R^2得分:", r2_score(y_test, y_pred))  # R方
```

```python

# 定义函数以评估模型性能
def evaluate(model, X, y):
    scores = cross_val_score(model, X, y, cv=5)
    print('交叉验证 R^2 分数：', np.mean(scores))
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    print('均方误差（MSE）：', mse)
    print('根均方误差（RMSE）：', rmse)
    r2 = r2_score(y, y_pred)
    print('R^2 分数：', r2)
```
