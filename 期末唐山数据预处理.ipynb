```python
import pandas as pd
```

```python
# 读取天气环境数据
weather_data = pd.read_csv('唐山天气数据11-22.csv',encoding='gbk')
```

```python
weather_data.head()  # 查看前几行数据
```

```python
weather_data.tail() #查看后几行数据
```

```python
# 检查缺失值
weather_data.isnull().sum()
```

```python
# 处理缺失值：删除缺失值或填充缺失值
weather_data.dropna()  # 删除包含缺失值的行
```

```python
weather_data.fillna(15)  # 填充缺失值，可以使用均值、中位数或自定义值等
```

```python
weather_data.duplicated()  # 检查重复值
```

```python
weather_data.drop_duplicates()  # 删除重复值所在的行
```

```python
weather_data.rename(columns={'最高温度': '平均温度'}, inplace=True)
```

```python
weather_data['平均温度'] = weather_data['平均温度'].astype(str)
```

```python
weather_data.to_csv('预处理后的唐山温度数据.csv', index=False) 
```
